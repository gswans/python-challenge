# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head(20)
```

```python
Total_Students = school_data_complete['student_name'].count()
Total_Schools = len(school_data_complete['school_name'].unique())
Total_Budget = sum(school_data_complete['budget'].unique())
Average_Math_Score = round(school_data_complete['math_score'].mean(), 6)
Average_Reading_Score = round(school_data_complete['reading_score'].mean(), 6)
Overall_Passing_Rate = round((Average_Math_Score + Average_Reading_Score)/2, 6) 
MP_Count = 0
RM_Score = school_data_complete['math_score']
for score_m in RM_Score:
    if score_m >= 70:
        MP_Count += 1
    Perc_MP = round((MP_Count/Total_Students)*100, 6)
RP_Count = 0
RR_Score = school_data_complete['reading_score']
for score_r in RR_Score:
    if score_r >= 70:
        RP_Count += 1
    Perc_RP = round((RP_Count/Total_Students)*100, 6)
#print(Total_Students)
#print(Total_Schools)
#print(Total_Budget)
#print(Average_Math_Score)
#print(Average_Reading_Score)
#print(Overall_Passing_Rate)
#print(Perc_MP)
#print(Perc_RP)
district_metrics = pd.DataFrame({'Total Schools':[Total_Schools], 
                                'Total Students':[Total_Students], 
                                'Total Budget':[Total_Budget], 
                                'Average Math Score':[Average_Math_Score], 
                                'Average Reading Score':[Average_Reading_Score], 
                                '% Passing Math':[Perc_MP], 
                                '% Passing Reading':[Perc_RP], 
                                'Overall Passing Rate':[Overall_Passing_Rate], 
                                }, index = None)
district_metrics
```
